```python
import numpy
import pandas as pd
%matplotlib inline
import seaborn
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Fremont Bridge Bike Counts.csv')
```

```python
times = pd.DatetimeIndex(data['date'])
Date = times.date
Hour = times.hour
Year = times.year
nb = data['fremont_bridge_nb']
sb = data['fremont_bridge_sb']

date2015 = []
hour2015 = []
nb2015 = []
sb2015 = []

for i in range(len(Year)):
    if Year[i] == 2015:
        date2015.append(Date[i])
        hour2015.append(Hour[i])
        nb2015.append(nb[i])
        sb2015.append(sb[i])

# Add fields for 2015 Date and hours and counts
# data['DATE2015'] = date2015
# data['HOUR2015'] = hour2015
# data['nb2015'] = nb2015
# data['sb2015'] = sb2015
```

```python
newdata = {'DATE2015':date2015, 'HOUR2015':hour2015, 'nb2015': nb2015, 'sb2015':sb2015}
newdata = pd.DataFrame(newdata)
```

```python
totalnb = newdata.groupby('HOUR2015')['nb2015'].sum()
# print totalnb
totalsb = newdata.groupby('HOUR2015')['sb2015'].sum()
# print totalsb
```

```python
nbplot = totalnb.plot(label = 'North Bound')
sbplot = totalsb.plot(label = 'South Bound')
seaborn.set()
handles, labels = nbplot.get_legend_handles_labels()
nbplot.legend(handles, labels)
nbplot.set_ylabel('Total Counts')
nbplot.set_title('Total Annual Traffic By Hour')
plt.savefig('BikesPerHour2015.png')
```

```python
nbmean = newdata.groupby('HOUR2015')['nb2015'].mean()
sbmean = newdata.groupby('HOUR2015')['sb2015'].mean()
```
